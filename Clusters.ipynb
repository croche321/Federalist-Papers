```python
#import docs

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file = "/content/drive/MyDrive/Colab Notebooks/data/HW4-data-fedPapers85.csv"

feddata = pd.read_csv(file)

feddata


```

```python
#Setting K-Means
from sklearn.cluster import KMeans

#Creating independent/dependent dataframes
columns_to_drop = ['author', 'filename']

X = feddata.drop(columns=columns_to_drop)

Y = feddata['author']
#K-means, udr k-means++ in order to get a greater distance between centroids, so the clusters are more clear.
kmeans = KMeans(n_clusters=4, init='k-means++')
kmeans.fit(X)
identified_clusters = kmeans.fit_predict(X)
X
```

```python
#Elbow Graph

distance=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(X)
    distance_iter = kmeans.inertia_
    distance.append(distance_iter)

number_clusters = range(1,7)
fig,ax=plt.subplots(figsize=(8,6))
ax.plot(number_clusters,distance)
ax.set(title='The Elbow title',
      xlabel='Number of clusters',
      ylabel='WCSS');

#We can see since the rate of decrease in WCSS starts to slow down after 4, the optimal kmean is 4.
```

```python
#Fitting Data to scatter plot
from sklearn.decomposition import PCA

data_with_clusters = X.copy()

data_with_clusters['Clusters'] = identified_clusters

#Reducing to two rows using PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(data_with_clusters)

#jittering data
jittered_data = pca_data + np.random.normal(0, .2, pca_data.shape)

#converting from array to df
pca_clustered_df = pd.DataFrame(jittered_data,columns=['pca1','pca2'])

#Scatter plot
plt.figure(figsize=(10, 8))
sns.scatterplot(x="pca2", y="pca1", hue=data_with_clusters['Clusters'], style=feddata['author'], data=pca_clustered_df)
plt.title('Federalist Papers Authors Clusters Visual')
plt.xlabel('')
plt.ylabel('')

#Crosstable for further visualization
table = pd.crosstab(feddata['author'], data_with_clusters['Clusters'])

table

#From this data, I came to the conclusion Madison is the writer of the disputed papers. In addition to this, the papers written by both Hamilton and Madison were primarily written in the style of Madison.
```
